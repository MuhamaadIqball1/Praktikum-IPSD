```python
import pandas as pd

# Membaca file CSV untuk data pelatihan dan pengujian
data_train = pd.read_csv('train_labels.csv')
data_test = pd.read_csv('test_features.csv')

# Menampilkan beberapa baris awal dari kedua dataset
print("Data Pelatihan:")
print(data_train.head())
print("\nData Pengujian:")
print(data_test.head())

# Melihat kolom yang ada pada kedua dataset
print("\nKolom pada Data Pelatihan:", data_train.columns)
print("\nKolom pada Data Pengujian:", data_test.columns)

# Menampilkan informasi umum terkait dataset (nilai kosong, tipe data, dll.)
print("\nInformasi Data Pelatihan:")
print(data_train.info())

print("\nInformasi Data Pengujian:")
print(data_test.info())

# Membandingkan statistik deskriptif dari kedua dataset
print("\nStatistik deskriptif Data Pelatihan:")
print(data_train.describe())

print("\nStatistik deskriptif Data Pengujian:")
print(data_test.describe())

# Mengecek jumlah nilai kosong pada kedua dataset
print("\nJumlah nilai kosong pada Data Pelatihan:")
print(data_train.isnull().sum())

print("\nJumlah nilai kosong pada Data Pengujian:")
print(data_test.isnull().sum())

# Mengecek kesamaan jumlah kolom atau nama kolom di kedua dataset
if set(data_train.columns) == set(data_test.columns):
    print("\nKolom pada kedua dataset identik.")
else:
    print("\nAda perbedaan kolom antara data pelatihan dan pengujian.")

# Membandingkan rata-rata nilai dari setiap kolom numerik
# Menyaring kolom numerik dari kedua dataset
kolom_numerik_train = data_train.select_dtypes(include=['number'])
kolom_numerik_test = data_test.select_dtypes(include=['number'])

# Mencari kolom numerik yang sama di kedua dataset
kolom_numerik_sama = kolom_numerik_train.columns.intersection(kolom_numerik_test.columns)

# Menghitung perbedaan rata-rata untuk kolom numerik yang sama
perbedaan_rata_rata = data_train[kolom_numerik_sama].mean() - data_test[kolom_numerik_sama].mean()

print("\nPerbedaan rata-rata nilai antara Data Pelatihan dan Pengujian:")
print(perbedaan_rata_rata)

```

```python
import pandas as pd

# Membaca file CSV untuk data train dan test
train_data = pd.read_csv('train_labels.csv')
test_data = pd.read_csv('test_features.csv')

# Menampilkan beberapa baris pertama dari kedua dataset
print("Data Train:")
print(train_data.head())
print("\nData Test:")
print(test_data.head())

# Memeriksa kolom yang terdapat di masing-masing dataset
print("\nKolom di Data Train:", train_data.columns)
print("\nKolom di Data Test:", test_data.columns)

# Menampilkan informasi dasar dari kedua dataset (missing values, tipe data, dll.)
print("\nInformasi Data Train:")
train_data.info()

print("\nInformasi Data Test:")
test_data.info()

# Memeriksa statistik deskriptif pada kedua dataset
print("\nStatistik Deskriptif Data Train:")
print(train_data.describe())

print("\nStatistik Deskriptif Data Test:")
print(test_data.describe())

# Memeriksa jumlah missing value di setiap dataset
print("\nJumlah missing value di Data Train:")
print(train_data.isnull().sum())

print("\nJumlah missing value di Data Test:")
print(test_data.isnull().sum())

# Mengecek kesamaan kolom antara kedua dataset
if set(train_data.columns) == set(test_data.columns):
    print("\nKolom di kedua dataset identik.")
else:
    print("\nAda perbedaan kolom antara dataset train dan test.")

# Membandingkan nilai rata-rata untuk setiap kolom numerik yang ada di kedua dataset
numeric_columns = train_data.select_dtypes(include=['number']).columns.intersection(test_data.select_dtypes(include=['number']).columns)
mean_difference = train_data[numeric_columns].mean() - test_data[numeric_columns].mean()

# --- Akhir perubahan ---

print("\nPerbedaan nilai rata-rata antara Data Train dan Data Test:")
print(mean_difference)
```
